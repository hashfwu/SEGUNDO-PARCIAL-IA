```python
import numpy as np
from IPython import display
import matplotlib.pyplot as plt
import os
from NeuralNetwork import TicTacToeNN
```

```python
# Inicialización del dataset
input_data = np.array(np.genfromtxt('data/input_data.csv', delimiter=','))
output_data = np.array(np.genfromtxt('data/output_data.csv', delimiter=','))  # Sin normalización adicional porque usamos tanh

# Normalización opcional de los datos
input_data = (input_data - np.mean(input_data)) / np.std(input_data)  # Normalización por si es necesario
output_data = (output_data - np.mean(output_data)) / np.std(output_data)
```

```python
INPUT_SIZE = 9
HIDDEN_SIZE = 18
OUTPUT_SIZE = 9
EPOCHS = 20000
LEARNING_RATE = 0.1
VIEW_RATE = 100
```

```python
NN = TicTacToeNN(INPUT_SIZE, HIDDEN_SIZE, OUTPUT_SIZE)
```

```python
NN.train(input_data, output_data, EPOCHS, LEARNING_RATE, VIEW_RATE, view_graph=True)
```
